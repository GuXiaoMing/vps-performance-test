```python
import pandas as pd
# from datetime import datetime as dt
import datetime
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
df = pd.read_csv("./vultr_statistics.csv", parse_dates=["time"])
```

```python
df = df[(df["time"] >= "2019-02-12 00:00:00") & (df["time"] < "2019-02-15 00:00:00")]
```

```python
avg_rtt_df = df[["time", "name", "avg_rtt"]].pivot(index="name", columns="time", values="avg_rtt").iloc[:, ::2]
bandwith_df = df[["time", "name", "bandwith"]].pivot(index="name", columns="time", values="bandwith").loc[:, ::2]
```

```python
for start_index in range(0, 16, 4):
    plt.figure()
    fig, axes = plt.subplots(nrows=2, ncols=1)
    # xticks = [datetime.datetime(2019, 2, 12) + datetime.timedelta(hours=x) for x in range(0, 24)]
    # xticks = [datetime.datetime(2019, 2, 12) + datetime.timedelta(hours=2 * x) for x in range(0, 24)]
    title = "Vultr Server rtt/ms & bandwith/MBps"
    avg_rtt_df.iloc[start_index: start_index + 4, :].transpose().plot(linewidth=0.5, title=title, ax=axes[0])
    lgd0 = axes[0].legend(bbox_to_anchor=(1.02,1.06), loc="upper left")
    # axes[0].set_xlim([xticks[0], xticks[-1]])
    axes[0].set_ylim([50,450])
    # axes[0].xaxis.set_ticks(xticks)
    # axes[0].xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%H"))
    axes[0].tick_params(axis='x', which='both', labelbottom=False)
    axes[0].xaxis.get_label().set_visible(False)

    bandwith_df.iloc[start_index: start_index + 4].transpose().plot(linewidth=0.5, ax=axes[1])
    lgd1 = axes[1].legend(bbox_to_anchor=(1.04,1.03), loc="upper left")
    axes[1].set_ylim([0,10])
    # axes[1].set_xlim([xticks[0], xticks[-1]])
    # axes[1].xaxis.set_ticks(xticks)
    # axes[1].xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%H"))
    # plt.xticks(fontsize=8, rotation=0)
    # plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=0, fontsize=7, ha="center")
    # plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=0, fontsize=7, ha="center")
    lgd1.remove()
    plt.savefig("./vultr_network_{}.png".format(start_index), dpi=500, bbox_extra_artists=(lgd0,), bbox_inches="tight")
```
